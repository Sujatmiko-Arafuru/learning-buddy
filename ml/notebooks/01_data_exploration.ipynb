```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys
import os

# Add path untuk import dari backend
sys.path.append(os.path.join(os.path.dirname(os.getcwd()), '../learning-buddy/backend'))

from db import collections

# Set style untuk plotting
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
```

## 1. Load Data


```python
# Load data dari MongoDB
if collections.get('student_progress'):
    progress_data = list(collections['student_progress'].find({}))
    df_progress = pd.DataFrame(progress_data)
    print(f"Student Progress: {len(df_progress)} records")
    print(df_progress.head())
    print(df_progress.info())
```

```python
# Load courses data
if collections.get('courses'):
    courses_data = list(collections['courses'].find({}))
    df_courses = pd.DataFrame(courses_data)
    print(f"Courses: {len(df_courses)} records")
    print(df_courses.head())
```
